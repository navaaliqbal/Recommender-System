```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
dataset = movies.merge(credits, on = 'title')
```

```python
dataset.info()
```

```python
# id,genres, keywords, overview, title, cast, crew

dataset = dataset[['movie_id','title','overview','genres','keywords','cast','crew']]
dataset.info()
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.dropna(axis = 0)
```

```python
dataset.duplicated().sum()
```

```python
import ast
def convert_genres(obj):
  list= []
  for i in ast.literal_eval(obj):
    list.append(i['name'])
  return list
```

```python
dataset['genres'] = dataset['genres'].apply(convert_genres)
```

```python
dataset.head()
```

```python
dataset['keywords'] = dataset['keywords'].apply(convert_genres)
dataset.head()
```

```python
dataset['cast'].iloc[0]
```

```python
def get_cast(obj):
  list = []
  cast = ast.literal_eval(obj)
  count = 0
  for i in cast:
    if count<3:
      list.append(i['name'])
  return list
```

```python
dataset['cast']= dataset['cast'].apply(get_cast)
```

```python
dataset.head()
```

```python
dataset['crew'].iloc[0]
```

```python
def get_job(obj,job):
  list = []
  crew = ast.literal_eval(obj)
  for i in crew:
    if i['job'] == job:
      list.append(i['name'])
  return list
```

```python
dataset['Director'] = dataset['crew'].apply(get_job,args = ("Director",))
```

```python
dataset['Writer'] = dataset['crew'].apply(get_job,args = ("Writer",))
dataset['Producer'] = dataset['crew'].apply(get_job,args = ("Producer",))
```

```python
del dataset['crew']
dataset.head()
```

```python
dataset['overview'] = dataset['overview'].apply(lambda x: x.split())
```

```python
dataset['genres'] = dataset['genres'].apply(lambda x: [i.replace(" ","") for i in x])
dataset['keywords'] = dataset['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
dataset['cast'] = dataset['cast'].apply(lambda x: [i.replace(" ","") for i in x])
dataset['Director'] = dataset['Director'].apply(lambda x: [i.replace(" ","") for i in x])
dataset['Producer'] = dataset['Producer'].apply(lambda x: [i.replace(" ","") for i in x])
dataset['Writer'] = dataset['Writer'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
dataset.head()
```

```python
dataset['tags'] = dataset['overview'] + dataset['genres'] + dataset['keywords'] + dataset['cast'] + dataset['Director'] + dataset['Producer']
```

```python
final_data = dataset[['movie_id', 'title', 'tags']]
```

```python
final_data.head()
```

```python
final_data.head()
```

```python
final_data['tags'] = final_data['tags'].apply(lambda x: ' '.join(x))
```

```python
final_data.head(10)
```

```python
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
stopwords = nltk.corpus.stopwords.words('english')
ps = PorterStemmer()
nltk.download('punkt')
```

```python
def transform_text(text):
  text = text.lower()
  text = ''.join([char for char in text if char not in string.punctuation])
  tokens = word_tokenize(text)
  final_tokens = [word for word in tokens if word not in stopwords]
  stemmed_tokens = [ps.stem(word) for word in final_tokens]
  final_text = ' '.join(stemmed_tokens)
  return final_text
```

```python
final_data['tags'] = final_data['tags'].apply(transform_text)
```

```python
final_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=5000)
result = vectorizer.fit_transform(final_data['tags'])
```

```python
result = result.toarray()
```

```python
result.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(result)
```

```python
similarity.shape
```

```python
indices = pd.Series(dataset.index, index=dataset['title']).drop_duplicates()
```

```python
list(enumerate(similarity[0]))
```

```python
def get_recommendation(movie_title):
  index = indices[movie_title]
  similarity_scores = similarity[index]
  movies_list = list(enumerate(similarity_scores))
  top_10 = sorted(movies_list,reverse= True, key = lambda x:x[1])[1:11]
  for i in top_10:
    print(final_data.iloc[i[0]].title)
```
